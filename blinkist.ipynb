```python
# redaing the csv file and converting it to a list 
df = pd.read_csv('urls.csv')

ind = 0
for i in df['book_url']:
    df['book_url'][ind] = i
    ind+=1
    
urls = list(df['book_url'])
print(len(urls))
```

```python
IS_NULL = 'Not Found'
lis_of_dic = []
count = 0
for x in urls:
    try:
        print(count)
        count+=1
        # load website 
        driver.get(x)
        time.sleep(5)
        try:
            title = driver.find_element(By.CSS_SELECTOR, "h1[data-test-id='book-title']").text
            print(title)
        except:
            title = IS_NULL
        try:
            subtitle = driver.find_element(By.CSS_SELECTOR, "p[class='text-p0 mb-6']").text
        except:
            subtitle = IS_NULL
        try:
            author = driver.find_element(By.CSS_SELECTOR, "h2[class='text-h5 font-bold mb-4 m:mb-6']").text
        except:
            author = IS_NULL
        dic = {}
        dic['Link'] = x
        dic['title'] = title
        dic['subtitle'] = subtitle
        dic['author'] = author


        # go read the story 
        try:
            read_text_button = driver.find_elements(By.CSS_SELECTOR, "a[data-test-id='read-button']")[1].click()
            time.sleep(5)

        except:
            print("no text available")


        cnt = 0
        while True:

                time.sleep(4)

                # find the element with the specified CSS selector
                page_content = driver.find_element(By.CSS_SELECTOR,"span[class='reader-content__text font-tisa-pro text-r2']").text

                # save the page content to the dictionary using a new key
                dic[f"chapter_{ cnt + 1}"] = page_content
                cnt+=1
                try:


                    # click the button with the specified data-test-id
                    button = driver.find_element(By.CSS_SELECTOR,"button[data-test-id='nextChapter']").click() 
                    # wait for the next page to load
                    time.sleep(4)
                
                except:
                # if the element is not found, break from the loop
                    print('finished')
                    break

        lis_of_dic.append(dic)
    except:
        print("error getting elements")

print("finished #####################")


```
